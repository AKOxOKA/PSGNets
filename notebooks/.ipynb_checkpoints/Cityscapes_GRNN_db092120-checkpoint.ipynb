# Dice

```python
import os
os.environ['CUDA_VISIBLE_DEVICES'] = ''
import tensorflow as tf
sess = tf.Session()
```

```python
hung = tf.load_op_library('/home/dbear/VisualVectorizingNetwork/vvn/ops/src/hungarian.so')
```

```python
def dice(log, lab):
    assert log.shape == lab.shape
    N = log.shape.as_list()[-1]
    numer = 2. * tf.reduce_sum(log * lab, axis=-1) + 1.
    denom = tf.reduce_sum(log, -1) + tf.reduce_sum(lab, -1) + 1.
    loss = 1. - (numer / denom)
    return tf.reduce_mean(loss)
    
```

```python
# lab = tf.zeros([5], dtype=tf.float32)
# log = 1. * tf.ones_like(lab)
lab = tf.constant([1.,1.,0.,0.], dtype=tf.float32)
log = (1. - lab) * 1.
loss = dice(log=log, lab=lab)
grad = tf.gradients(loss, log)[0]

L, G = sess.run([loss, grad])
print(L, G)

print(sess.run(log))
```

```python
labels = tf.constant([1.,1.,0.,0.,1.], dtype=tf.float32)
labels = tf.stack([labels, 1.-labels], axis=0)
logits = 1. - labels
logits = 0.001 * logits
print(sess.run(logits), sess.run(labels))
```

```python
labels = tf.constant([1.]*1 + [0.]*19, dtype=tf.float32)
labels = tf.stack([labels, 1.-labels], axis=0)
logits = 1. - labels
logits = logits * tf.constant([1/19., 1.])[:,tf.newaxis]
print(sess.run(logits), sess.run(labels))
```

```python
labels = tf.constant([0,0,5,3,1,2,7,7,5,4,3,3,6,0,0], dtype=tf.int32)
labels = tf.cast(tf.one_hot(labels, depth=8, axis=0), tf.float32)
# logits = 1. - labels
logits = tf.concat([0.1*tf.ones_like(labels[0:1]), tf.zeros_like(labels[1:])], axis=0)
sess.run(labels)
```

```python
logs = 0.01 * logits
for step in range(20):
    logs = tf.minimum(tf.nn.relu(logs), 1.)
    logs /= tf.reduce_max(logs, axis=1, keepdims=True) + 1e-8
    logs /= tf.reduce_sum(logs, axis=0, keepdims=True)  
    loss = dice(logs, labels)

    print("loss and logs before step", step, sess.run([loss, logs]))
    grad = tf.gradients(loss, logs)[0]
    logs -= grad

    
```

```python
print(sess.run(logs[0]))
print(sess.run(labels[0]))
```
